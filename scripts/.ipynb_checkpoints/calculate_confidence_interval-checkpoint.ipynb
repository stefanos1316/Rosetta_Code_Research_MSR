```python
import pandas as pd
import seaborn as sns
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import scipy.stats
import re
import math
import matplotlib.colors as mcolors
from os import environ


#All environmental variables are located here
#index = environ['myvar']
#numberOfTasks = int(environ['noTasks']) + 0
numberOfTasks = 25
#confideceInterval = environ['confidenceInterval']
confideceInterval = 95
z = 1.960
#numberOfLanguages = int(environ['noLanguages']) + 0
numberOfLanguages = 14
totalFields = numberOfTasks*numberOfLanguages
#tasksNames = environ['allTasksNames']
tasksNames=['ackermann','array-concat', 'call-obj-method', 'classes','concurrent-comp.', 'exponentiation-operator','factorial','file-I/O','function-composition' ,'huffman', 'inherit,-multi.','inherit.-single','json','lzw-compression','numerical-integration','palindrome','regex','bubble-sort','insertion-sort','merge-sort','quick-sort','selection-sort','synchronous-concurrency','url-decoding','url-encoding']
#tasksNames = tasksNames.split(",")
#ileLocation = environ['fileLocation']
#fullPath = fileLocation + "Normalized_EDP_" + index + ".txt"


#Initializing Arrays
langaugesEDP=[""  for x in range(totalFields)]
ProgrammingLanguages=[""  for x in range(totalFields)]
ProgrammingTasks=[""  for x in range(totalFields)]
langaugeArray=['C#', 'C', 'C++', 'Go', 'JavaScript', 'Java', 'Perl', 'PHP', 'Python', 'R', 'Ruby', 'Rust', 'Swift', 'VB.Net']
#taskArray=['ackermann','array-concat', 'call-obj-method', 'classes','concurrent-comp.','factorial','file-I/O', 'huffman', 'inherit,-multi.','inherit.-single','json','palindrome','regex','bubble-sort','insertion-sort','merge-sort','quick-sort','selection-sort','url-decoding','url-encoding']
counter=0      

#getCPU_Architecture=re.split('/', fullPath)


#Read files and fill tables with ata
#with open(fullPath) as f:
with open('/home/sgeorgiou/GitHub/Rosetta_Code_Research_MSR/Reports/server_processor/volatile_tasks/GraphData/Normalized_EDP_1.txt') as f:
    for line in f:
        line = line.replace(":","")
        line = line.replace(",","")
        line = line.replace("-1","np.nan")
        data = re.split('\s+', line)
        data.pop(0)
        for i in range(numberOfLanguages):
            langaugesEDP = np.append(float(data[i]))
            #counter += 1
            
correctlySortedEDP=[[0 for x in range(numberOfLanguages)] for y in range(numberOfTasks)] 
counter = 0
for i in range(numberOfTasks):
    for j in range (numberOfLanguages):
        correctlySortedEDP[i][j] = langaugesEDP[counter]
        counter += 1 
        
        
#Currectly all the values are stored in correctlySortedEDP array, now we will calculate the mean for each task
taskMeanValueArray = [0 for x in range(numberOfTasks)] 
denominators = [0 for x in range(numberOfTasks)] 
counter = 0
for i in range(numberOfTasks):
    totalValue = 0
    denominator = 0
    for j in range (numberOfLanguages):
         if correctlySortedEDP[i][j] != -1:
            totalValue += correctlySortedEDP[i][j]
            denominator += 1
    denominators[counter] = denominator        
    taskMeanValueArray[counter] = totalValue/denominator
    counter += 1
    
#Now we will calculate the variance between our values
variance = [0 for x in range(numberOfTasks)] 
counter = 0
for i in range(numberOfTasks):
    getTotal = 0
    for j in range (numberOfLanguages):
        getTotal += math.pow((taskMeanValueArray[i] - correctlySortedEDP[i][j]),2) 
    variance[counter] =  getTotal / denominators[i]
    #print(variance[counter])
    #print(denominators[i])
    counter += 1
    
standardDeviation = [0 for x in range(numberOfTasks)]    
for i in range(numberOfTasks):   
    standardDeviation[i] = math.sqrt(variance[i])
    #print(standardDeviation[i])
    
#This array includes all the itnervals for the specific means    
intervalArray = [0 for x in range(numberOfTasks)] 
for i in range(numberOfTasks):  
    intervalArray[i] = 1.960 * (standardDeviation[i]/math.sqrt(denominators[i]))
    
for i in range(numberOfTasks):  
    print('Condidence interval for Task:', tasksNames[i] ,' Mean:' , str(round(taskMeanValueArray[i],2)) , "Interval: +/-" , str(round(intervalArray[i],2)) )
```
